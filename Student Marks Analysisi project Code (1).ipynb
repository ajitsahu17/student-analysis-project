### Step 2: Python Code

```python
import csv

# Step 1: Read data from CSV
def read_data(filename):
    students = []
    with open(filename, "r") as file:
        reader = csv.DictReader(file)
        for row in reader:
            students.append({
                "Name": row["Name"],
                "Math": int(row["Math"]),
                "Science": int(row["Science"]),
                "English": int(row["English"])
            })
    return students

# Step 2: Calculate average marks
def calculate_average(student):
    return (student["Math"] + student["Science"] + student["English"]) / 3 

# Step 3: Generate report
def generate_report(students):
    report = []
    for student in students:
        avg = calculate_average(student)
        status = "Pass" if avg >= 50 else "Fail"
        report.append({
            "Name": student["Name"],
            "Average": round(avg, 2),
            "Status": status
        })
    return report

# Step 4: Save report to CSV
def save_report(report, filename):
    with open(filename, "w", newline="") as file:
        writer = csv.DictWriter(file, fieldnames=["Name", "Average", "Status"])
        writer.writeheader()
        writer.writerows(report)

# Step 5: Main function
def main():
    students = read_data(r"G:\My Drive\NetTech India\Python\Python Project\Project 1\Student.csv")
    report = generate_report(students)
    save_report(report, "report2.xlsx")
    print("Report generated successfully!")

# main program 
if __name__ == "__main__":
    main()
```
